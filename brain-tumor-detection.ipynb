## 1.3 Data Processing

### 1.3.1 Create Target Indicator

```python
# Initialize data and tumor indicator
image_data = []
image_filename = []
tumor_ind = []

# Append data and tumor indicator
for img_file in tumor_yes_files + tumor_no_files:
    # Get image filepath
    if img_file in tumor_yes_files:
        img_filepath = os.path.join(tumor_yes_dir, img_file)
    else:
        img_filepath = os.path.join(tumor_no_dir, img_file)
    
    # Fetch image
    img = Image.open(img_filepath)
    
    # Append image
    image_data.append(img)
    
    # Append image filename
    image_filename.append(img_file)
    
    # Append tumor indicator
    if img_file in tumor_yes_files:
        tumor_ind.append(1)
    else:
        tumor_ind.append(0)
    
# Check if append operations were performed correctly
total_file_num = len(tumor_yes_files) + len(tumor_no_files)
assert len(image_data) == total_file_num, 'Images were not appended correctly'
assert len(image_filename) == total_file_num, 'Image filenames were not appended correctly'
assert len(tumor_ind) == total_file_num, 'Tumor indicator were not appended correctly'

# Print success message
print('Image dataset and tumor indicator have been successfully formed.\n')
print('Total images retrieved: ', len(image_data), f' = Images w Tumor ({len(tumor_yes_files)}) + Images w/o Tumor ({len(tumor_no_files)})')
print('Total indicators retrieved: ', str(len(tumor_ind)), f' = Indicator Yes ({str(sum(tumor_ind))}) + Indicator No ({str(len(tumor_ind) - sum(tumor_ind))})')
```

### 1.3.2 Split into Train, Valid & Test Datasets

```python
# Split dataset into train and test (70% vs 30%)
seed_num = 69
X_train, X_test, y_train, y_test = train_test_split(image_data, tumor_ind, test_size=0.3, random_state=seed_num)

# Split test dataset into valid and test (15% vs 15%)
X_test, X_valid, y_test, y_valid = train_test_split(X_test, y_test, test_size=0.5, random_state=seed_num)

# Check if splits were performed correctly
assert len(X_train) + len(X_test) + len(X_valid) == len(image_data), 'Images were not split correctly'
assert len(y_train) + len(y_test) + len(y_valid) == len(tumor_ind), 'Indicators were not split correctly'

# Print success message
print(f'Total images: Training = {len(X_train)}, Valid = {len(X_valid)}, Test = {len(X_test)}.', '\n')

# Create directory
train_dir = os.path.join(cd, 'brain_tumor_dataset\\train')
valid_dir = os.path.join(cd, 'brain_tumor_dataset\\valid')
test_dir = os.path.join(cd, 'brain_tumor_dataset\\test')

# Create directory for images with tumor and those without
for directory in [train_dir, valid_dir, test_dir]:
    for sub_directory in [directory, directory + '\\1', directory + '\\0']:
        if os.path.exists(sub_directory):
            shutil.rmtree(sub_directory)
        
        os.mkdir(sub_directory)

# Store images into respective folders
for img_folder, ind_folder in zip([X_train, X_valid, X_test], [y_train, y_valid, y_test]):
    for img, ind in zip(img_folder, ind_folder):
        # Get image index
        img_index = image_data.index(img)

        # Locate image filename
        img_filename = image_filename[img_index]
        img_filename = img_filename.replace('.jpeg', '.jpg')

        # Convert to RGB channels
        img_rgb = img.convert('RGB')
        
        # Determine directory
        if img in X_train:
            directory = train_dir
        elif img in X_valid:
            directory = valid_dir
        else:
            directory = test_dir

        # Save in either '1' or '0' folder
        if ind == 1:
            img_rgb.save(os.path.join(directory + '\\1', img_filename))
        else:
            img_rgb.save(os.path.join(directory + '\\0', img_filename))
        
# Print success message
print(f'Train dataset has {str(sum(y_train))} images with tumor, and {str(len(y_train) - sum(y_train))} images without tumor.')
print(f'Valid dataset has {str(sum(y_valid))} images with tumor, and {str(len(y_valid) - sum(y_valid))} images without tumor.')
print(f'Test dataset has {str(sum(y_test))} images with tumor, and {str(len(y_test) - sum(y_test))} images without tumor.')
```

### 1.3.3 Data Preprocessing

```python
# No. of subprocesses to use for data loading, where 0 means all processes
num_workers = 0

# No. of samples per batch to load
batch_size = 1

# Convert data to a normalized torch.FloatTensor
train_transform = transforms.Compose([
    transforms.Resize((64, 64)), # 64 x 64 square images
#     transforms.Grayscale(), # convert to greyscale
    transforms.RandomHorizontalFlip(), # randomly flip the image horizontally
    transforms.RandomRotation(10), # randomly tilt with 10 degrees angle
    transforms.ToTensor(), # convert to tensors
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)) # mean & std dev for RGB
#     transforms.Normalize((0.5), (0.5)) # mean & std dev for B&W
])

test_transform = transforms.Compose([
    transforms.Resize((64, 64)), # 64 x 64 square images
#     transforms.Grayscale(), # convert to greyscale
    transforms.ToTensor(), # convert to tensors
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)) # mean & std dev for RGB
#     transforms.Normalize((0.5), (0.5)) # mean & std dev for B&W
])

# Import the train, valid and test datasets
train_data = datasets.ImageFolder(train_dir, transform=train_transform)
valid_data = datasets.ImageFolder(valid_dir, transform=train_transform)
test_data = datasets.ImageFolder(test_dir, transform=test_transform)

# Define samplers for obtaining training and validation datasets
train_idx, valid_idx = list(range(len(train_data))), list(range(len(valid_data)))
train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)

# Prepare data loaders (combine dataset and sampler)
train_loader = torch.utils.data.DataLoader(
    train_data, 
    batch_size=batch_size,
    sampler=train_sampler, 
    num_workers=num_workers
)

valid_loader = torch.utils.data.DataLoader(
    valid_data, 
    batch_size=batch_size, 
    sampler=valid_sampler, 
    num_workers=num_workers
)

test_loader = torch.utils.data.DataLoader(
    test_data, 
    batch_size=batch_size, 
    num_workers=num_workers
)

# Specify the loaders
loaders = {'train': train_loader, 'valid': valid_loader, 'test': test_loader}

# Print out number of images stored in respective folders
print('No. of train images: ', len(train_data))
print('No. of valid images: ', len(valid_data))
print('No. of test images: ', len(test_data))
```

# 2. Self-Built CNN Model

## 2.1 Model Architecture

```python
# Set loss function
criterion = nn.CrossEntropyLoss()

# Set optimizer
def get_optimizer(model):
    # Adam optimizer
    optimizer = optim.Adam(
        model.parameters(), 
        lr=0.00001
    )
    
    # SGD optimizer
#     optimizer = optim.SGD(
#         model.parameters(), 
#         lr=0.01, 
#         momentum=0.9
#     )

    return optimizer


# Use pytorch's default weight initialization
def default_weight_init(m):
    # Retrieve default parameters
    reset_parameters = getattr(m, 'reset_parameters', None)
    
    # Reset parameters
    if callable(reset_parameters):
        m.reset_parameters()
        
        
# Set custom weight initialization
def custom_weight_init(m):
    # Find class name
    class_name = m.__class__.__name__
    
    # For each Linear layer in a model, initialize weights = 1 / sqrt(n)
    if class_name.find('Linear') != -1:
        # Get the number of the inputs
        n = m.in_features
        
        # Calculate the standard deviation
        y = (1 / np.sqrt(n))
        
        # Initialize weights using N(0, n^(-0.5))
        m.weight.data.normal_(0, y)
        
        # Set all biases to 0
        m.bias.data.fill_(0)    
        

# Define the CNN architecture
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()     
        # Convolutional layer (sees 64x64x3 image tensor)
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
        
        # Convolutional layer (sees 32x32x16 tensor)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        
        # Convolutional layer (sees 16x16x32 tensor)
        self.conv3 = nn.Conv2d(32, 64, 3, padding=1)
        
        # Convolutional layer (sees 8x8x64 tensor)
        self.conv4 = nn.Conv2d(64, 128, 3, padding=1)
        
        # Max pooling layer
        self.pool = nn.MaxPool2d(2, 2)
        
        # Linear layer (128 * 4 * 4 -> 1000)
        self.fc1 = nn.Linear(128 * 4 * 4, 1000)
        
        # Linear layer (1000 -> 500)
        self.fc2 = nn.Linear(1000, 500)
        
        # Linear layer (500 -> 2) as there are 2 classes in total
        self.fc3 = nn.Linear(500, 2)
        
        # Dropout layer (p=0.25)
        self.dropout = nn.Dropout(0.25)      
    
    def forward(self, x):
        
        # Add sequence of convolutional and max pooling layers
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = self.pool(F.relu(self.conv4(x)))
        
        # Flatten image input
        x = x.view(x.size(0), -1)
        x = self.dropout(x) # add dropout layer
        
        # Add 1st hidden layer, with relu activation function
        x = F.relu(self.fc1(x))
        x = self.dropout(x) # add dropout layer
        
        # Add 2nd hidden layer, with relu activation function
        x = F.relu(self.fc2(x))
        x = self.dropout(x) # add dropout layer
        
        # Add 3rd hidden layer, without relu activation function
        x = self.fc3(x)
        
        return x

    
# Instantiate the CNN
model = CNN()

# Apply weight initialization
model.apply(custom_weight_init) # model.apply(default_weight_init)

# Determine if GPU is available
use_cuda = torch.cuda.is_available()
print('GPU is available.' if use_cuda else 'GPU is not available.')

# Move tensors to GPU if CUDA is available
if use_cuda:
    model.cuda()#.half()
```

## 2.2 Model Training

```python
# Define train function
def train(n_epochs, loaders, model, optimizer, criterion, use_cuda, model_name):
    # Initialize tracker for minimum validation loss
    min_valid_loss = np.Inf
    
    # Initialize dataframe to store results
    epoch_ls = []
    train_loss_ls = []
    valid_loss_ls = []
    
    for epoch in range(1, n_epochs+1):
        # Initialize training and validation loss
        train_loss = 0.0
        valid_loss = 0.0
        
        # Set model to training mode
        model.train()
        
        # Loop through training dataset
        for batch_idx, (data, target) in enumerate(loaders['train']):
            
            # Move model training to GPU
            if use_cuda:
                data, target = data.cuda(), target.cuda()
       
            # Clear the gradients of all optimized variables
            optimizer.zero_grad()
            
            # Perform forward pass
            output = model(data)
            
            # Calculate the batch loss
            loss = criterion(output, target)
            
            # Perform backward pass
            loss.backward()
            
            # Perform a single optimization step (parameter update)
            optimizer.step()
            
            # Record the average training loss
            train_loss = train_loss + ((1 / (batch_idx + 1)) * (loss.data - train_loss))

        # Set model to evaluation mode
        model.eval()
        
        # Loop through validation dataset
        for batch_idx, (data, target) in enumerate(loaders['valid']):
            # Move model training to GPU
            if use_cuda:
                data, target = data.cuda(), target.cuda()

            # Perform forward pass
            output = model(data)
            
            # Calculate the batch loss
            loss = criterion(output, target)
            
            # Record the average validation loss 
            valid_loss = valid_loss + ((1 / (batch_idx + 1)) * (loss.data - valid_loss))

        # Print training & validation statistics 
        print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(str(int(epoch)), train_loss, valid_loss))
        
        # Append model results
        epoch_ls += [str(int(epoch))]
        train_loss_ls += [np.round(train_loss.cpu().data.numpy(), 6)]
        valid_loss_ls += [np.round(valid_loss.cpu().data.numpy(), 6)]
        
        # Save model at the filepath stored in save_path if the validation loss has decreased
        if valid_loss <= min_valid_loss:
            print('Validation loss decreased ({:.6f} --> {:.6f}). Saving model ...'.format(min_valid_loss, valid_loss))
            
            # Save the model
            torch.save(model.state_dict(), model_name)
            
            # Update the new minimum loss
            min_valid_loss = valid_loss
    
    # Compile results in dataframe format
    model_results = ({
        'Epoch': epoch_ls, 
        'Train_Loss': train_loss_ls, 
        'Valid_Loss': valid_loss_ls
    })
    model_results_df = pd.DataFrame(model_results)
        
    return model, model_results_df


# Set number of epochs
num_epochs = 200

# Train the model
model, model_results_df = train(
    num_epochs, 
    loaders, 
    model, 
    get_optimizer(model), 
    criterion, 
    use_cuda, 
    'self_built_cnn_model.pt'
)

# Plot training and validation loss
ax = model_results_df.plot.line()
ax.legend(loc='upper right')
plt.title('Self-Built CNN Model Results')
plt.xlabel('Epoch')
plt.show()
```

## 2.3 Model Validation

```python
# Define test validation
def test(loaders, model, criterion, use_cuda):

    # Initialize test loss and accuracy
    test_loss = 0
    correct = 0
    total = 0
    
    # Set model to evaluation mode
    model.eval()
    
    # Loop through test dataset
    for batch_idx, (data, target) in enumerate(loaders['test']):
        # Move model training to GPU
        if use_cuda:
            data, target = data.cuda(), target.cuda()
            
        # Perform forward pass
        output = model(data)

        # Calculate the batch loss
        loss = criterion(output, target)

        # Record the average test loss 
        test_loss = test_loss + ((1 / (batch_idx + 1)) * (loss.data.item() - test_loss))
        
        # Convert output probabilities to predicted class
        pred = output.data.max(1, keepdim=True)[1]
        
        # Compare predictions to true indicator
        correct += np.sum(np.squeeze(pred.eq(target.data.view_as(pred))).cpu().numpy())
        total += data.size(0)
        
    # Calculate accuracy
    test_accuracy = 100 * correct / total
        
    # Print results            
    print('Test Loss: {:.6f}\n'.format(test_loss), '\n')
    print('Test Accuracy: %2d%% (%2d/%2d)' % (test_accuracy, correct, total))
    
    return test_loss, test_accuracy

# load the model that got the best validation accuracy
model.load_state_dict(torch.load('self_built_cnn_model.pt'))

# Perform model validation
model_test_loss, model_accuracy = test(loaders, model, criterion, use_cuda)
```

# 3. ResNet Model (Transfer Learning)

## 3.1 Model Architecture

```python
# Specify model architecture
resnet_model = models.resnet50(weights='ResNet50_Weights.DEFAULT')

# Freeze training for all "features" layers (convolutional layers)
for param in resnet_model.parameters():
    param.requires_grad = False
    
# Add last linear layer (n_inputs -> 2 classes), which automatically have requires_grad=True
resnet_model.fc = nn.Sequential(
    nn.Linear(2048, 1024),
    nn.ReLU(inplace=True),
    nn.Dropout(0.25),
    nn.Linear(1024, 2)
)

# Move model training to GPU
if use_cuda:
    resnet_model = resnet_model.cuda()
    
# Show model architecture
print('ResNet Model Architecture:')
resnet_model
```

## 3.2 Model Training

```python
# Train the model
resnet_model, resnet_model_results_df = train(
    num_epochs, 
    loaders, 
    resnet_model, 
    get_optimizer(resnet_model), 
    criterion, 
    use_cuda, 
    'resnet_pretrained_model.pt'
)

# Plot training and validation loss
ax = resnet_model_results_df.plot.line()
ax.legend(loc='upper right')
plt.title('ResNet Pre-Trained Model (Transfer-Learning) Results')
plt.xlabel('Epoch')
plt.show()
```

## 3.3 Model Validation

```python
# load the model that got the best validation accuracy
resnet_model.load_state_dict(torch.load('resnet_pretrained_model.pt'))

# Perform model validation
resnet_model_test_loss, resnet_model_accuracy = test(loaders, resnet_model, criterion, use_cuda)
```
